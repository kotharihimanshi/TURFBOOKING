```python
#Himanshi Kothari
#22BTRSN020
# Turf class to represent each turf
class Turf:
    def __init__(self, name, availability, price ):
        self.name = name
        self.availability = availability
        self.price = price
        
   

# Initialize a list to store turf objects
turfs = []

turfs.append(Turf("Turf A", True, 500.0 ))
turfs.append(Turf("Turf B", True, 700.0 ))
turfs.append(Turf("Turf C", True, 900.0 ))
turfs.append(Turf("Turf D", True, 1200.0 ))


# Function to display available turfs
def display_available_turfs():
    print("\nAvailable Turfs:")
    for index, turf in enumerate(turfs):
        if turf.availability:
            print(f"{index+1}. {turf.name} - ${turf.price} ")
            
            

# Function to check turf availability
def check_turf_availability():
    display_available_turfs()
    selected_turf = int(input("Select a turf to check availability: ")) - 1
    if 0 <= selected_turf < len(turfs):
        turf = turfs[selected_turf]
        if turf.availability:
            print(f"{turf.name} is available.")
        else:
            print(f"{turf.name} is not available.")
    else:
        print("Invalid turf selection.")
        





# Function to book a turf
def book_turf():
    display_available_turfs()
    selected_turf =int(input("Select a turf to book: ")) - 1
    if 0 <= selected_turf < len(turfs):
        turf = turfs[selected_turf]
        if turf.availability:
            date = input("Enter booking date (YYYY-MM-DD): ")
            time = input("Enter booking time (HH:MM AM/PM): ")

            duration = float(input("Enter booking duration (in hours): "))

            print("\nTurf booking summary:")
           
            print("Turf:", turf.name)
            print("Date:", date)
            print("Time:", time)
            print("Duration:", duration, "hours")
            print("Total Price:", turf.price * duration)

            confirmation = input("Confirm booking (yes/no): ")
            if confirmation.lower() == 'yes':
                turf.availability = False
                print("Turf booked successfully!")
                print("thank you")
            else:
                print("Booking canceled.")
        else:
            print("Turf is not available for booking.")
    else:
        print("Invalid turf selection.")

# Main menu
while True:
    print("\n----- Turf Booking System -----")
    print("1. Check Turf Availability")
    print("2. Book Turf")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")
   

    if choice == '1':
        input("Enter your name:")
        int(input("Enter your phone number:"))
        check_turf_availability()
    elif choice == '2':
        input("Enter your name:")
        int(input("Enter your phone number:"))
        book_turf()
    elif choice == '3':
        print("thank you!!")
        break
    else:
        print("Invalid choice. Please try again.")
```
